```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import accuracy_score
from scipy.stats import mode
from sklearn.naive_bayes import GaussianNB
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
train.head()
```

```python
test.head()
```

```python
train = train.drop(columns=["Unnamed: 0", "id"])
label_encoders = {}
for col in train.select_dtypes(include="object").columns:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    label_encoders[col] = le

X_train = train.drop(columns=["satisfaction"])
y_train = train["satisfaction"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)

kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
agglo = AgglomerativeClustering(n_clusters=2)

kmeans_labels = kmeans.fit_predict(X_scaled)
agglo_labels = agglo.fit_predict(X_scaled)

def best_accuracy(true_labels, cluster_labels):
    labels = np.zeros_like(cluster_labels)
    for i in range(2):
        mask = (cluster_labels == i)
        labels[mask] = mode(true_labels[mask], keepdims=True).mode[0]
    return accuracy_score(true_labels, labels)

kmeans_acc = best_accuracy(y_train, kmeans_labels)
agglo_acc = best_accuracy(y_train, agglo_labels)

kmeans_acc, agglo_acc
```
